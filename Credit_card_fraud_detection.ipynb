This dataset is highly imbalanced

0 --> Normal transactions

1 --> Fraud transactions

```python
#separate the data for analysis
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
 fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

Here we get the difference between the legit and fraud data values

```python
#under-sampling
legit_sample = legit.sample(n=492)
```

```python
#concate two dataset to get balanced dataset
balanced_dataset = pd.concat([legit_sample, fraud],axis= 0)
```

```python
balanced_dataset.head()
```

```python
balanced_dataset.tail()
```

```python
balanced_dataset.shape
```

```python
balanced_dataset['Class'].value_counts()
```

```python
balanced_dataset.groupby('Class').mean()
```

###**Train-Test-Split**

```python
x = balanced_dataset.drop(columns='Class',axis=1)
y = balanced_dataset['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

### **Model Building**

**LR Model**

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(x_train,y_train)
```

```python
y_pred_lr = model_lr.predict(x_test)
y_pred_lr
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("The accuracy score is: ",accuracy_score(y_test,y_pred_lr)*100)
print("The confusion matrix is: \n",confusion_matrix(y_test,y_pred_lr))
print("The classification report is: \n",classification_report(y_test,y_pred_lr))
```

**DTC Model**

```python
from sklearn.tree import DecisionTreeClassifier
model_dtc = DecisionTreeClassifier()
model_dtc.fit(x_train,y_train)
```

```python
y_pred_dtc = model_dtc.predict(x_test)
y_pred_dtc
```

```python
print("The accuracy score is: ",accuracy_score(y_test,y_pred_dtc)*100)
print("The confusion matrix is: \n",confusion_matrix(y_test,y_pred_dtc))
print("The classification report is: \n",classification_report(y_test,y_pred_dtc))
```

**RFC Model**

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier()
model_rf.fit(x_train,y_train)
```

```python
y_pred_rf = model_rf.predict(x_test)
y_pred_rf
```

```python
print("The accuracy score is: ",accuracy_score(y_test,y_pred_rf)*100)
print("The confusion matrix is: \n",confusion_matrix(y_test,y_pred_rf))
print("The classification report is: \n",classification_report(y_test,y_pred_rf))
```

**SVC Model**

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(x_train,y_train)
```

```python
y_pred_svc = model_svc.predict(x_test)
y_pred_svc
```

```python
print("The accuracy score is: ",accuracy_score(y_test,y_pred_svc)*100)
print("The confusion matrix is: \n",confusion_matrix(y_test,y_pred_svc))
print("The classification report is: \n",classification_report(y_test,y_pred_svc))
```

**KNN Model**

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier()
model_knn.fit(x_train,y_train)
```

```python
y_pred_knn = model_knn.predict(x_test)
y_pred_knn
```
